```python
common_accouunt_type = dataset['Account Type'].value_counts().idxmax()
count = dataset['Account Type'].value_counts().max()

print(f'Most common account type is: {common_accouunt_type} Account')
print(f'Most common account type count: {count}')
```

```python
most_common_country = dataset['Country'].value_counts().idxmax()
count = dataset['Country'].value_counts().max()

print(f'Most common country to have an account: {most_common_country}')
print(f'Most common country count: {count}')
```

```python
dataset['City'].value_counts()
```

```python
eldest_person = dataset['Age'].max()
youngest_person = dataset['Age'].min()
average_age = dataset['Age'].mean().astype(int)

print(f'Eldest person to have an account: {eldest_person}')
print(f'Youngest person to have an account: {youngest_person}')
print(f'Average age to have an account: {average_age}')
```

```python
dataset['Account Creation Date'] = pd.to_datetime(dataset['Account Creation Date'])
dataset['Last Active Date'] = pd.to_datetime(dataset['Last Active Date'])
```

```python
highest_balance = dataset['Balance (USD)'].max()
print('Highest balance:', highest_balance)

highest_balance_row = dataset[dataset['Balance (USD)']== highest_balance]
highest_balance_row
```

```python
lowest_balance = dataset['Balance (USD)'].min()
print('Lowest balance:', lowest_balance)

lowest_balance_row = dataset[dataset['Balance (USD)']== lowest_balance]
lowest_balance_row
```

```python
highest_debt = dataset['Debt (USD)'].max()
print('Highest debt:', highest_debt)

highest_debt_row = dataset[dataset['Debt (USD)']== highest_debt]
highest_debt_row
```

```python
lowest_debt = dataset['Debt (USD)'].min()
print('lowest debt:', lowest_debt)

lowest_debt_row = dataset[dataset['Debt (USD)']== lowest_debt]
lowest_debt_row
```

```python
credit_score = dataset['Credit Score'].astype(int)
highest_credit_score= credit_score.max()
print('Highest credit score:', highest_credit_score)

highest_credit_score_row = dataset[dataset['Credit Score'] == highest_credit_score]
highest_credit_score_row
```

```python
lowest_credit_score= credit_score.min()
print('Lowest credit score:',lowest_credit_score)

lowest_credit_score_row = dataset[dataset['Credit Score'] == lowest_credit_score]
lowest_credit_score_row
```

```python
highest_income = dataset['Income Level (USD Annual)'].max()
print(highest_income)

highest_income_row = dataset[dataset['Income Level (USD Annual)'] == highest_income]
highest_income_row
```

```python
lowest_income = dataset['Income Level (USD Annual)'].min()
print(lowest_income)

lowest_income_row = dataset[dataset['Income Level (USD Annual)'] == lowest_income]
lowest_income_row
```

```python
sorted_data = dataset.sort_values(by= 'Balance (USD)', ascending= False)
top50_unique = sorted_data.drop_duplicates(subset= 'Full Name').head(50)
plt.figure(figsize= (15, 9))
sns.barplot(y= 'Full Name', x= 'Balance (USD)', data= top50_unique)
plt.show()
```

```python
types_of_account= dataset['Account Type'].value_counts().sort_values()
types_of_account.plot(kind= 'barh')
plt.title('Types of Account')
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sns.countplot(x= 'Country', data= dataset, order= dataset['Country'].value_counts().index)
plt.title('Customer Distribution by Country')
plt.xticks(rotation= 90)
plt.ylabel('Number of Customers')
plt.show()

# This figure shows the number of customers from each country.
```

```python
top10 = dataset.groupby(['Country', 'City'])['Balance (USD)'].mean().sort_values(ascending=False).head(10).reset_index()
top10['Location'] = top10['Country'] + " - " + top10['City']
sns.barplot(y= 'Location', x= 'Balance (USD)', data= top10)
plt.show()
# Brazil - Rio de Janeiro is the richest location followed by Sao Paulo.
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(x= 'Country', hue= 'Risk Category', data=dataset)
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize= (7, 5))
gender= dataset['Gender'].value_counts().plot(kind= 'pie', autopct= '%1.2f%%', startangle= 90)
plt.title('Male and Female')
plt.ylabel('')
plt.legend()
plt.show()
```

```python
total_amt_by_gender = dataset.groupby('Gender')['Balance (USD)'].sum()
print(total_amt_by_gender)
sns.barplot(x= total_amt_by_gender.index, y= total_amt_by_gender)
plt.show()
```

```python
plt.figure(figsize= (7, 5))
plt.hist(dataset['Debt (USD)'], bins= 35)
plt.xlabel('Debt (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Debt (USD)')
plt.show()
```

```python
sns.countplot(x= 'Risk Category', data= dataset)
plt.xlabel('Risk Category')
plt.ylabel('Number of Customers')
plt.title('Customer Distribution by Risk Category')
plt.show()
```

```python
dataset['KYC Status'].value_counts().plot(kind= 'pie', autopct= '%1.2f%%', startangle= 360, explode= [0, 0.04])
plt.title('Proportion of customers by KYC Status')
plt.axis('equal')
plt.ylabel('')
plt.show()
```
